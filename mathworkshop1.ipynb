```python
def number_group(number):
  if number > 0:
    return "Positive"
  elif number == 0:
    return "Zero"
  else:
    return "Negative"
print(number_group(10)) #Should be Positive
print(number_group(0)) #Should be Zero
print(number_group(-5)) #Should be Negative
```

```python

def fib():
    n = int(input("Enter the number of terms: "))
    a = 0
    b = 9
    print("Fibonacci sequence:")
    for i in range(n):
        print(a, end=" ")
        c = a + b
        a = b
        b = c
fib()
```

```python

def myPow(x, n):
    result = 1
    count = 0
    while count < n:
        result = result * x
        count += 1
    print(result)
x = int(input("Enter positive integer: "))
n = int(input("Enter power: "))
myPow(x,n)
```

```python

def SumDigits(n):
    total = 0
    while n > 0:
        digit = n % 10
        total = total +  digit
        n = n // 10
    print(total)
num = int(input("Enter a number: "))
SumDigits(num)
```

```python

def squaredSum(arr):
    total = 0
    for num in arr:
        total = total + (num * num)
    print(total)

squaredSum([2,3,4])
```

```python
def sumDigits(num):
    total = 0
    for i in range(num+1):
        total = total + i
    print(total)

num = int(input("Enter a number: "))
sumDigits(num)
```
