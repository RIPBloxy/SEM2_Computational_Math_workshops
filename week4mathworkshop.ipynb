```python
import matplotlib.pyplot as plt
car_colour=['Red','Silver','Blue','White','Green']
number_seen=[123,230,78,193,12]
myexplode = [0.3, 0, 0,0,0]
plt.pie(number_seen,labels=car_colour, explode=myexplode, shadow = True)
plt.title("Car color and number seem ")
plt.xlabel("Car colour ")
plt.ylabel("Number seen")
plt.show()
```

```python
import matplotlib.pyplot as plt
car_colour=['Red','Silver','Blue','White','Green']
number_seen=[123,230,78,193,12]
myexplode = [0.3, 0, 0,0,0]
plt.pie(number_seen,labels=car_colour, explode=myexplode, shadow = True)
plt.legend(title = "Car Colours:")
plt.title("Car color and number seem ")
plt.xlabel("Car colour ")
plt.ylabel("Number seen")
plt.show()
```

```python
import matplotlib.pyplot as plt
car_colour=['Red','Silver','Blue','White','Green']
number_seen=[123,230,78,193,12]
myexplode = [0.3, 0, 0,0,0]
plt.pie(number_seen,labels=car_colour, explode=myexplode, shadow = True,colors = car_colour)
plt.legend(title = "Car Colours:")
plt.title("Car color and number seem ")
plt.xlabel("Car colour ")
plt.ylabel("Number seen")
plt.show()
```
