The following table shows the different vehicles and the distance covered by them in the span of 5 days. Display the following table in line plot.

a. Plot “days” in X- axis and “distance covered” in Y- axis.

b. Title of the table “bike details in Plot”

```python
import matplotlib.pyplot as plt
import math
days = [1, 2, 3, 4, 5]
enfield = [50, 80, 70, 40]
honda   = [80, 20, 20, 20]
yamaha  = [70, 20, 60, 30]
ktm     = [20, 60, 60, 60]
plt.plot(days, enfield, marker='o', label="Enfield")
plt.plot(days, honda, marker='o', label="Honda")
plt.plot(days, yamaha, marker='o', label="Yamaha")
plt.plot(days, ktm, marker='o', label="KTM")
plt.title("Bike Details in Plot")
plt.xlabel("Days")
plt.ylabel("Distance (kms)")
plt.legend()
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import math
days = [1, 2, 3, 4, 5]
enfield = [50, 40, 70, 80, 20]
honda   = [80, 20, 20, 50, 60]
yamaha  = [70, 20, 60, 40, 60]
ktm     = [40, 20, 30, 70, 60]

plt.subplot(4,1,1)
plt.plot(days, enfield)
plt.title("Enfield")

plt.subplot(4,1,2)
plt.plot(days, honda)
plt.title("Honda")

plt.subplot(4,1,3)
plt.plot(days, yamaha)
plt.title("Yamaha")

plt.subplot(4,1,4)
plt.plot(days, ktm)
plt.title("KTM")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import math
x = []
y = []
i = -3 * math.pi
while i <= 3 * math.pi:
    x.append(i)
    y.append(math.cos(i))
    i += 0.1

plt.plot(x, y)
plt.title("y = cos(x)")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import math
x = []
y = []
theta = 0
while theta <= 2 * math.pi:
    x.append(4 * math.cos(theta) + 2)
    y.append(4 * math.sin(theta) + 2)
    theta += 0.1

plt.plot(x, y)
plt.title("Circle")
plt.grid()
ax = plt.gca()
ax.set_aspect('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt
import math
x1 = []
y1 = []
x2 = []
y2 = []
theta = 0
while theta <= 2 * math.pi:
    x1.append(4 * math.cos(theta) + 2)
    y1.append(4 * math.sin(theta) + 2)

    x2.append(3 * math.cos(theta) + 2)
    y2.append(5 * math.sin(theta) + 2)

    theta += 0.1

plt.plot(x1, y1, label="Circle")
plt.plot(x2, y2, label="Ellipse")
plt.legend()
plt.title("Circle and Ellipse")
plt.grid()
ax = plt.gca()
ax.set_aspect('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt
import math
x_vals = []
y_vals = []
p = 0
q = 0
a = 0.5
x = -5
while x <= 5:
    y = 4 * a * (x - q)**2 + p
    x_vals.append(x)
    y_vals.append(y)
    x += 0.1

plt.plot(x_vals, y_vals)
plt.title("Parabola")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid()
plt.show()
```
